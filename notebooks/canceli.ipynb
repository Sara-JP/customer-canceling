```python
# do some preprocessing 
train_data.isna().sum()
```

```python
def My_prepro(train_data, test_data, x):
    if x in train_data.columns:
        train_data = train_data.drop(x, axis=1)
    if x in test_data.columns:
        test_data = test_data.drop(x, axis=1)
    return train_data, test_data

# Assuming train_data and test_data are already defined

prepro_cols = ['CancelTime', 'BillID', 'TicketID', 'UserID', 'From', 'To', 'VehicleType', 'VehicleClass', 'HashPassportNumber_p', 'HashEmail', 'BuyerMobile', 'NationalCode']

for col in prepro_cols:
    train_data, test_data = My_prepro(train_data, test_data, col)
```

```python
from datetime import datetime

def days_between(d1, d2):
    #d1 = datetime.strptime(d1, "%Y-%m-%d")
    #d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1))

train_data['ndays']= days_between(train_data['Created'],train_data['DepartureTime'])
test_data['ndays']= days_between(test_data['Created'],test_data['DepartureTime'])
```

```python
train_data['ndays']=train_data['ndays'].dt.days
test_data['ndays']=test_data['ndays'].dt.days

train_data
```

```python
from sklearn.preprocessing import LabelEncoder

lable_encoder = LabelEncoder()


def My_LabelEncoder(x):
    train_data[x]= lable_encoder.fit_transform(train_data[x])
    test_data[x]= lable_encoder.transform(test_data[x])
categorical_cols = ['Male','Vehicle','TripReason']
for i in categorical_cols:
    My_LabelEncoder(i)
    
 

# do some preprocessing 
trip_months= [6, 7, 8, 9]
train_data['DepartureTime_day']= train_data['DepartureTime'].dt.day
train_data['DepartureTime_month']= train_data['DepartureTime'].dt.month
train_data['DepartureTime_year']= train_data['DepartureTime'].dt.year

train_data['is_vacation']= [month in trip_months for month in train_data['DepartureTime_month']]


test_data['DepartureTime_day']= test_data['DepartureTime'].dt.day
test_data['DepartureTime_month']= test_data['DepartureTime'].dt.month
test_data['DepartureTime_year']= test_data['DepartureTime'].dt.year

test_data['is_vacation']= [month in trip_months for month in test_data['DepartureTime_month']]

test_data = test_data.drop('DepartureTime', axis =1)
train_data = train_data.drop('DepartureTime', axis =1)

test_data = test_data.drop('Created', axis =1)
train_data = train_data.drop('Created', axis =1)
train_data
```

```python
mapping = {False: 0,
           True : 1}

train_data['is_vacation'] = train_data['is_vacation'].map(mapping)
test_data['is_vacation'] = test_data['is_vacation'].map(mapping)


test_data
```

```python
# modeling
# modeling 
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier

model = AdaBoostClassifier(estimator=RandomForestClassifier(n_estimators=2),n_estimators=10,random_state=20,learning_rate=1) 


x_train =train_data.drop(['Cancel'] , axis = 1)
model.fit(x_train.to_numpy(), train_data['Cancel'])
```
